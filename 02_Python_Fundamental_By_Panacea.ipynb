### `Task 1` Python Loops

### `Definition of Loops`
loop is a fundamental control structure used to repeat a specific block of code multiple times until a certain condition is met. Loops are essential for automating repetitive tasks, such as processing every item in a dataset or performing iterative calculations

### `Most Used Loops:`
1. For Loop
2. While Loop

### `For Loop`
`Basic Syntax`

for item in list:
  print(item)
Here,
1.  for: The keyword that starts the loop
2.  item: A variable name you choose (it acts as a placeholder for the current element being processed).
3. in: A keyword used to link the variable to the collection.
4. list: The collection you want to travel through (e.g., a List) a String, a Range, or a Column in your data set
5. : (Colon): Crucial. This signals the start of the "loop body

```python
# Example of a for loop
molecules = [ "CCOC", "OCCOC", "CCCCOOCC", "OCCOCC" ]

for mol in molecules:
    print(mol [2])
```

```python
# Example of while loop

count = 1

while count <= 5:
    print(f"Molecule number {count}")
    count += 1  # This is the "Update" step
```

## Python Tuples

### `Definition of Tuples:`
a tuple is a built-in data type used to store collections of data. They are very similar to lists, but with one critical distinction: immutability. Once a tuple is created, you cannot change, add, or remove its elements.

### `Used Cases in Chem Informatics`
- Atomic Coordinates and Vectors
- Dictionary Keys for Molecular Properties
- Representing Chemical Bonds


```python
# A simple tuple representing a point in 3D space
atom_coordinates = (12.5, 45.2, -3.1)
print(atom_coordinates)
print(type(atom_coordinates)) # Type Check
```

```python
# A simple tuple representing a point in 3D space
atom_coordinates = (12.5, 45.2, -3.1)
print(atom_coordinates)
print(type(atom_coordinates)) # Type Check

# To 'modify' a tuple, you typically create a new one.
# Method 1: Convert to a list, modify, then convert back to a tuple

# Let's say you want to change the second element (45.2 to 50.0)
coordinates_list = list(atom_coordinates)
coordinates_list[1] = 50.0
new_atom_coordinates = tuple(coordinates_list)
print(f"Original tuple: {atom_coordinates}")
print(f"Modified (new) tuple: {new_atom_coordinates}")

# Method 2: Create a new tuple by combining parts of the old tuple and new elements

# Let's say you want to add an element to the end of the tuple
# You can't directly add, but you can create a new tuple with the original elements + new ones
new_element = 10.0
expanded_atom_coordinates = atom_coordinates + (new_element,)
print(f"Expanded (new) tuple: {expanded_atom_coordinates}")

# Let's say you want to 'remove' an element
# You can create a new tuple with slices of the original, excluding the element you want to 'remove'
# For example, removing the second element (index 1)
removed_atom_coordinates = atom_coordinates[:1] + atom_coordinates[2:]
print(f"Tuple with element 'removed' (new tuple): {removed_atom_coordinates}")
```

## Conditional Statements and
##`What is Conditional Statement `
In Python, a conditional statement is used to determine the execution flow of a program based on whether a specific condition evaluates to True or False. It allows the code to "make decisions" and execute different blocks of logic depending on the input or state of the data.

## `Types of Statements`
- if
- else
- else if elese

```python
# If, elif, else  conditional Statement by making bmi calculator

# 1. Get user input
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

# 2. Calculate BMI
bmi = weight / (height ** 2)

print(f"Your BMI is: {bmi:.2f}")

# 3. Use Conditional Statements to categorize the result
if bmi < 18.5:
    print("Category: Underweight")
elif 18.5 <= bmi < 25:
    print("Category: Normal weight")
elif 25 <= bmi < 30:
    print("Category: Overweight")
else:
    # This catches any BMI 30 or higher
    print("Category: Obese")
```

## Logical Operator
In Python, logical operators are used to combine multiple conditional statements or to reverse the logic of a single condition. They allow you to create more complex decision-making rules in your code beyond a simple True or False
- AND
- OR
- NOT



```python
# Understanding the logical operators by evaluating lipinski rule

# Molecular data for a sample compound
molecule_name = "Compound X"
mw = 450      # Molecular Weight
logp = 4.2    # Lipophilicity
hbd = 2       # Hydrogen Bond Donors
hba = 7       # Hydrogen Bond Acceptors

# 1. Individual Boolean Checks
mw_ok = mw <= 500
logp_ok = logp <= 5
hbd_ok = hbd <= 5
hba_ok = hba <= 10

# 2. Using 'and' to check if ALL rules are met perfectly
perfect_match = mw_ok and logp_ok and hbd_ok and hba_ok

# 3. Using 'or' to allow for one violation (The practical Lipinski limit)
# We count violations; if violations <= 1, it passes.
violations = 0
if not mw_ok: violations += 1
if not logp_ok: violations += 1
if not hbd_ok: violations += 1
if not hba_ok: violations += 1

# Final logic gate using 'and' and 'or'
if perfect_match:
    print(f"{molecule_name} passed all Lipinski rules perfectly!")
elif violations == 1:
    print(f"{molecule_name} passed (with 1 minor violation).")
else:
    print(f"{molecule_name} failed validation (More than 1 violation).")
```

## Introduction to Numpy and Pandas
NumPy is the foundation for scientific computing in Python. It provides the ndarray (n-dimensional array), which is a powerful data structure that is much faster and more memory-efficient than standard Python lists.
Pandas is built on top of NumPy and is used for data manipulation and analysis. It introduces the DataFrame, which is essentially a programmable spreadsheet (similar to an Excel table or a SQL table).

```python
# Numpy introduction by a simple task
import numpy as np

# A list of Binding Energies (kcal/mol) for 3 ligands
energies = np.array([-7.2, -8.5, -6.1])

# Convert kcal/mol to kJ/mol (multiply all by 4.184)
energies_kj = energies * 4.184

print(energies_kj)
# Output: [-30.1248 -35.564  -25.5224]
```

```python
import pandas as pd

# Data from a virtual screening run
data = {
    'Ligand_ID': ['ZINC_01', 'ZINC_02', 'ZINC_03', 'ZINC_04'],
    'MW': [450.234, 620.110, 380.550, 410.890],
    'Docking_Score': [-9.21, -7.45, -4.82, -8.11],
    'LogP': [3.2, 5.8, 2.1, 4.5]
}

df = pd.DataFrame(data)

# 'Beautify' the data: Round values and filter for Lipinski compliance (MW <= 500)
df['MW'] = df['MW'].round(1)
hits = df[df['MW'] <= 500].sort_values(by='Docking_Score')

print(hits)
```
