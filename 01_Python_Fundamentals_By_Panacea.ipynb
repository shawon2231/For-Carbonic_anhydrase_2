### `Task 01:` Basic Syntax (Printing a text and Variables)
### `Why are we Interesd in printing text and what we will do?`
In bioinformatics, DNA, RNA, and protein sequences are treated as strings of characters rather than numerical values.
### `Definition of Variable`
a variable is essentially a reserved memory location to store values.

### `Variable Naming Convension`
1. Use snake_case: Join lowercase words with underscores (e.g., dna_sequence, sample_id, mean_growth_rate)
2. Be Descrivtive in names (eg: sequence_1 or seq_1 not s1)
3. Do not Start With numbers (eg: Do not use 1st_sequence)
4. Never use reserved key words

### `What we actually do : The String Operation`
1. Printing The Text (STRINGS)
2. String Concatenation
3. Finding The Length of The String
4. Changing The case of The String

```python
 # Printing a text: [Syntax: Function_name("Text Here")] !Do not forget to use "" for Plain text (Strings)

print("Welcome To Panacea") # Here, Function name = print("")
```

```python
# Declearing a variable [Syntax: Variable_name = value]
My_name = "Here is your trainer Pritom Kundu"
print(My_name)
```

```python
# Strings Concatenation Joining a String
seq_1 = "ACCGTAA"
seq_2 = "ATGTTC"
# Concatenation Syntax: String 1 + String 2
seq_3 = seq_1 +  seq_2
print(seq_3)
```

```python
# Creating Concatenation with Space
seq_1 = "ACCGTAA" # Define seq_1 based on previous cell
seq_2 = "ATGTTC" # Define seq_2 based on previous cell
seq_4 = seq_1 + " " + seq_2
print(seq_4)
```

```python
# Finding the length of a string [Means: How Many charecters] Using function len()
seq_fasta ="ATGGCCATGCATTATGCATCGGCATCGACATCGATGCAAATTCGATCGATGCCGTA" # Decleared a variable
# Lets Check the Fasta Length
len(seq_fasta)
```

```python
# Changing The case of a String
# 01 Changing Upper to lower cases
fasta_lower = seq_fasta.lower()
print(fasta_lower)

# 02 Changing my lower case to upper case
fasta_2 = "atcggctattcgaata"
fasta_upper = fasta_2.upper()
print(fasta_upper)
```

### `Task 2:`Exploring The Data Types
### `What is data types`
In Python, data types are classifications that tell the interpreter how you intend to use a specific piece of data.

### `Mostly Used Data Types are:`
- Numeric [int, Float, complex]
- Sequence Types [text, strings]
- Mapping Type
- Boolean [True False]
- Binary [0 1]


```python
# Exploring Data types : Numeric
# Syntax: main_function(child_function(argument)) Here main_function is print(), child_function is type()

item_1 = 45
print(type(item_1)) # This is a nested function : Means Function inside another Function
```

```python
# Exploring Data types : Text, Steings
item_2 = "Panacea"
print(type(item_2))
```

```python
# Exploring Data types : Mapping Type [Means Dictionary]
item_3 = {"Name": "Panacea", "Location": ""}
print(type(item_3))
print(item_3)
```

```python
# Exploring Data types : Boolean [True False]
intem_4 = True
print(type(intem_4))
```

```python
# Exploring Data types : Binary [0, 1]
item_5 = 0
print(type(item_5))
```

### `Task 03:` Type Casting Conversion of Data Types
### `Definition:`

In Python, Type Casting (also known as Type Conversion) is the process of converting a value from one data type to another.

### `Most used Cases`
1. String to Numeric
2. Numeric to String
3. Collection Casting (The "Unique/Mutable" Pattern)


```python
# String to Numeric
user_age = "25"
print(type(user_age))
user_age

# Convertung The String value to Numeric
user_age = int(user_age)
print(type(user_age))
user_age
```

```python
#  String to Numeric
my_age = 28
print(type(my_age))
my_age

# Conversion
my_age = str(my_age)
print(type(my_age))
my_age
```

```python
# Collection Casting (The "Unique/Mutable" Pattern) [Conversion of List to Set]
# A list with duplicate IDs
raw_ids = [101, 102, 101, 105, 102]
print(type(raw_ids))

# Converting to set
unique_ids = set(raw_ids)
print(type(unique_ids))
unique_ids
```
